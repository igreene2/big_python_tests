## Linear Support Vector Machine

#### Generate artificial gaussian data.

```python
# Generate binary class dataset
center = [5,5]
n_samples = 100
X1 = np.random.normal(0,5,(int(n_samples/2),2)) - center
X2 = np.random.normal(0,5,(int(n_samples/2),2)) + center
X = np.concatenate((X1,X2),0)
Y = np.ones((n_samples,))
Y[:int(n_samples/2)] = -1

plt.figure()
plt.scatter(X1[:,0],X1[:,1],marker='x')
plt.scatter(X2[:,0],X2[:,1],marker='o')
plt.show()
```

1.Train a support vector machine on the data with $C=1$ and draw a line that separates o and x. Mark each of the support vectors separately and clearly (e.g. by putting a circle around the point).

Note: You can use LinearSVC or SVC from scikit-learn. If you use SVC, be sure to use the linear kernel as follows: SVC(kernel='linear'...).

```python
from sklearn.svm import SVC
clf = SVC( C = 1.0, kernel = "linear" )
clf.fit(X,Y)
print(clf.score(X,Y))
spv = clf.support_vectors_ 
w = clf.coef_
d = clf.dual_coef_
b = clf.intercept_
wl = -w[0,0]/w[0,1]
print(spv)
print(w)
print(d)
plt.figure()
plt.scatter(X1[:,0],X1[:,1],marker='x')
plt.scatter(X2[:,0],X2[:,1],marker='o')
plt.plot(X[:,0], wl*X[:,0]+b, linestyle='dotted', color = 'g')
plt.scatter(spv[:,0], spv[:,1],s=120, facecolors='none', edgecolors='r')
plt.show()
```

2.Draw a line that separates the data for 8 different $C$ ($10^{-3}$~$10^5$). Also, plot the number of support vectors vs. $C$ (plot x-axis on a log scale). How does the number of support vectors change as $C$ increases and why does it change like that? 

```python
C_group = np.logspace(-3,5,9)
w_group = np.zeros(9)
b_group = np.zeros(9)
sv_num = np.zeros(9)

for i in range(9):
    clf = SVC( C = C_group[i], kernel = "linear")
    clf.fit(X,Y)
    w = clf.coef_
    b_group[i] = clf.intercept_
    w_group[i] = -w[0,0]/w[0,1]
    sv_num[i] = np.sum(clf.n_support_)
    plt.figure()
    print('C =', (C_group[i]))
    plt.scatter(X1[:,0],X1[:,1],marker='x')
    plt.scatter(X2[:,0],X2[:,1],marker='o')
    plt.plot(X[:,0], w_group[i]*X[:,0]+b_group[i], linestyle='dotted')
    plt.show()
    
    
```

```python
plt.figure()
plt.plot(C_group, sv_num)
plt.title("support vector numbers vs. C")
plt.xscale("log")
plt.show()

print("The number of support vectors is decreased as C increase, for the penalty of the slackness is greater, thus less vectors were seleceted as support vectors")
```

7.Train one-vs-rest classifier on the train data to classify subjects 0~37. Use a support vector machine with $C=0.01$. Plot a bar graph of each subject's accuracy. What is the overall accuracy of the classifier?

Note: Use sklearn.multiclass.OneVsRestClassifier which allows you to convert any appropriate estimation function into a multi-class classifier.

```python
from sklearn.multiclass import OneVsRestClassifier
ovrc = OneVsRestClassifier(SVC(C = 0.01, kernel = "linear"))
ovrc.fit(F_train_p, Y_train)
acc_train = ovrc.score(F_train_p, Y_train)
acc_test = ovrc.score(F_test_p, Y_test)

print("overall accuracy for train data ", acc_train)
print("overall accuracy for test data ", acc_test)
```

```python
Y_train_predict = ovrc.predict(F_train_p)
Y_test_predict = ovrc.predict(F_test_p)

acc_sum_train = Y_train_predict - Y_train
acc_sum_test = Y_test_predict - Y_test

acc_ovc_train = np.zeros(38)
acc_ovc_test = np.zeros(38)


for i in range(38):
    acc_ovc_train[i] = (40 - np.count_nonzero(acc_sum_train[40*i:40*(i+1)])) / 40
    loc = np.where(Y_test == i)[0]
    num_t = loc.shape[0]
    acc_ovc_test[i] = (num_t - np.count_nonzero(acc_sum_test[loc]))/num_t
```
